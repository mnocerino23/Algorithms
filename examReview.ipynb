```python
input = [1, 4, 5, 9, 20, 10, 3, 15, 17, 20, 23]
```

```python
min = []
def getSmallestDistance(p, low, high):
  if low == high:
    return min[-1]
  min.append(abs(p[high]-p[high-1]))
  rest = getSmallestDistance(p, low, high-1)
  if rest < min[-1]:
    min.append(rest)
```

```python
getSmallestDistance(input, 0, 10)
```

```python
def largestNegativeSubarray(A, low, high):
  if containsPositive(A, low, high) == False:
    return high - low
  firstNegative = getFirstNegative(A, low, high)
  secondNegative = getSecondNegative(A, low, high, firstNegative)
  print(firstNegative, secondNegative)
  largestNegativeSubarray(A, firstNegative, secondNegative)
  largestNegativeSubarray(A, secondNegative, high)
```

```python
def getFirstNegative(A, low, high):
  for i in range(low, high+1):
    if A[i] < 0:
      return i
  return -1
```

```python
def getSecondNegative(A, low, high, firstNegative):
  for i in range(low, high +1):
    if A[i] < 0 and i != firstNegative:
      return i
  return -1
```

```python
def containsPositive(A, low, high):
  for i in range(low, high+1):
    if A[i] >= 0:
      return True
  return False
```
